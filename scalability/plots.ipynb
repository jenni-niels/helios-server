```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
steps = [1,5,10,20,50,100]

results = {
    "1Q_10C": {"tally": [1.635,6.809, 12.245,24.941, 74.598, 117.532], "decrypt":[1.118,0.783,0.762,0.907,0.941,0.823]},
    "4Q_4C":  {"tally": [2.363,12.681,23.901,51.406, 109.890,220.862], "decrypt":[1.190,1.286,1.118,1.157,1.391,1.151]},
    "4Q_10C": {"tally": [7.011,30.744,58.231,103.157,276.290,526.171], "decrypt":[3.958,3.427,3.308,3.615,3.003,3.025]},
    "10Q_4C": {"tally": [6.940,30.576,59.577,101.206,282.109,588.512], "decrypt":[3.386,3.513,2.960,2.285,3.660,2.742]},
}


```

```python
plt.figure(figsize=(8,6))
for k,v in results.items():
    plt.plot(steps, results[k]["tally"], label=k)
plt.legend()
plt.title("Vote Tallying Time")
plt.xlabel("# Voters")
plt.ylabel("Time (s)")
plt.savefig("figures/tally_time.png", dpi=300)
plt.show()


plt.figure(figsize=(8,6))
for k,v in results.items():
    plt.plot(steps, results[k]["decrypt"], label=k)
plt.legend()
plt.title("Vote Decryption Time")
plt.xlabel("# Voters")
plt.ylabel("Time (s)")
plt.savefig("figures/decrypt_time.png", dpi=300)
plt.show()
```
